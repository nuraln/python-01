# Area Plot

### 1. Area Plot - Persebaran kasus menurut wilayah periode 2008-2016
Menampilkan persebaran jumlah kasus tertinggi dan terendah di wilayah-wilayah pada periode 2008-2016 melalui area plot.

Value : jumlah kasus

Borought : Wilayah

```python
df_bor=df['value'].groupby(df['borough']).sum()

plt.title('Jumlah Kasus Tiap Wilayah pada Tahun 2008-2016', fontdict={'fontname':'Arial', 'fontsize': 20, 'fontweight': 'bold', 'verticalalignment': 'bottom', 'y': 3.1})
plt.fill_between(df_bor.index, df_bor.values, alpha=1, color='#DFA67B')

plt.xticks(rotation=90, fontsize=9)
plt.xlabel('Wilayah', fontsize=15,fontweight = 'semibold', fontname='Arial')
plt.ylabel('Jumlah Kasus', fontsize=15, fontweight = 'semibold', fontname='Arial')

plt.show()
```

### 2. Area Plot - Persebaran kasus pada tahun 2008
Untuk melihat detail persebaran kasus pada tahun 2008 menggunakan area plot

```python
# filter data for year 2008
df_2008 = df[df['year'] == 2008]
```

```python
# calculate total crimes per borough in 2008
df_bor_2008 = df_2008['value'].groupby(df_2008['borough']).sum()

# create plot
plt.title('Jumlah Kasus Tiap Wilayah pada Tahun 2008', fontdict={'fontname':'Arial', 'fontsize': 20, 'fontweight': 'bold', 'verticalalignment': 'bottom', 'y': 1.05})
plt.fill_between(df_bor_2008.index, df_bor_2008.values, alpha=1, color='#ABC4AA')

plt.xticks(rotation=90, fontsize=9)
plt.xlabel('Wilayah', fontsize=15, fontweight='semibold', fontname='Arial')
plt.ylabel('Jumlah Kasus', fontsize=15, fontweight='semibold', fontname='Arial')

plt.show()
```

### 3. Area Plot - Persebaran kasus pada tahun 2016
Untuk melihat detail persebaran kasus pada tahun 2016 menggunakan area plot

```python
# filter data for year 2016
df_2016 = df[df['year'] == 2016]

# calculate total crimes per borough in 2016
df_bor_2016 = df_2016['value'].groupby(df_2016['borough']).sum()

# create plot
plt.title('Jumlah Kasus Tiap Wilayah pada Tahun 2016', fontdict={'fontname':'Arial', 'fontsize': 20, 'fontweight': 'bold', 'verticalalignment': 'bottom', 'y': 1.05})
plt.fill_between(df_bor_2016.index, df_bor_2016.values, alpha=1, color='#F4B183')

plt.xticks(rotation=90, fontsize=9)
plt.xlabel('Wilayah', fontsize=15, fontweight='semibold', fontname='Arial')
plt.ylabel('Jumlah Kasus', fontsize=15, fontweight='semibold', fontname='Arial')

plt.show()
```

Analisis dari plot ini dapat dilihat dari beberapa aspek berikut:

1. Wilayah dengan jumlah kasus tertinggi: Dari plot ini, dapat dilihat bahwa wilayah Westminster memiliki jumlah kasus terbanyak dari semua wilayah, diikuti oleh Southwark dan Lambeth. Sementara itu, wilayah City of London memiliki jumlah kasus paling sedikit.

2. Perbedaan jumlah kasus antar wilayah: Dari plot ini, dapat dilihat bahwa ada perbedaan jumlah kasus yang signifikan antara wilayah-wilayah tertentu. Beberapa wilayah memiliki jumlah kasus yang jauh lebih tinggi daripada wilayah lainnya.

3. Tren perubahan jumlah kasus dari waktu ke waktu: Plot ini tidak memperlihatkan tren perubahan jumlah kasus dari waktu ke waktu, karena hanya memperlihatkan jumlah kasus untuk seluruh periode 2008-2016 secara keseluruhan. Namun, plot ini dapat digunakan untuk membandingkan jumlah kasus antar wilayah pada tahun yang sama.


# Line Plot

### 1. Line Plot - Kasus Major Category di Westminster Tahun 2008-2016
Untuk melihat Major Category apa yang merupakan kejahatan tertinggi dan terendah di Westminster pada tahun 2008-1016

```python
# Creating a variable to store major case data for Westminster.
df_westminster = df[(df['borough']=='Westminster')].groupby('major_category')['value'].sum()

# create plot
plt.title('Jumlah Kasus Major Category di Wilayah Westminster pada Tahun 2008-2016', fontdict={'fontname':'Arial', 'fontsize': 20, 'fontweight': 'bold', 'verticalalignment': 'bottom', 'y': 1.05})
plt.fill_between(df_westminster.index, df_westminster.values, alpha=0.5, color='orange', facecolor='lightblue')
df_westminster.plot(kind='line',color='navy')
plt.xticks(rotation=90, fontsize=15)
plt.xlabel('Major Category', fontsize=15, fontweight='semibold', fontname='Arial')
plt.ylabel('Jumlah Kasus', fontsize=15, fontweight='semibold', fontname='Arial')
plt.show()
```

### 2. Line Plot - Kasus Minor Category di Westminster Tahun 2008-2016
Untuk melihat Minor Category apa yang merupakan kejahatan tertinggi dan terendah di Westminster pada tahun 2008-1016

```python
# Creating a variable to store minor case data for Westminster.
df_westminster = df[(df['borough']=='Westminster')].groupby('minor_category')['value'].sum()

# create plot
plt.title('Jumlah Kasus Minor Category di Wilayah Westminster pada Tahun 2008-2016', fontdict={'fontname':'Arial', 'fontsize': 20, 'fontweight': 'bold', 'verticalalignment': 'bottom', 'y': 1.05}}
plt.fill_between(df_westminster.index, df_westminster.values, alpha=0.5, color='orange', facecolor='#b8621b')
df_westminster.plot(kind='line',color='#b8621b')
plt.xticks(rotation=90, fontsize=10)
plt.xlabel('Minor Category', fontsize=15, fontweight='semibold', fontname='Arial')
plt.ylabel('Jumlah Kasus', fontsize=15, fontweight='semibold', fontname='Arial')
plt.show()
```

Dari grafik garis tersebut, kita dapat melihat bahwa mayoritas kasus kejahatan minor di Westminster adalah pencurian. Sedangkan untuk kejahatan mayor, yang paling umum adalah kejahatan kekerasan (violent crimes), diikuti dengan pencurian.

Tentu saja, analisis lebih lanjut diperlukan untuk menggali lebih dalam mengenai faktor-faktor apa yang mempengaruhi tren ini, apakah terdapat peran faktor ekonomi, kebijakan keamanan, atau faktor-faktor lainnya.

# Histogram

### Histogram - Jumlah Borough Area Berdasarkan Jumlah Kriminalitas

```python
plt.figure(figsize=(12, 5))
df.groupby(['borough'])[['value']].sum().hist(bins=20, color='#27e1c1')
plt.title('Jumlah Borough District Area Berdasarkan Jumlah Kriminalitasnya', fontdict={'fontname':'Arial', 'fontsize': 20, 'fontweight': 'bold', 'verticalalignment': 'bottom', 'y': 1.05})
plt.xlabel('Frekuensi Kriminal',fontsize=15, fontweight='semibold', fontname='Arial')
plt.ylabel('Jumlah borough', fontsize=15, fontweight='semibold', fontname='Arial')
plt.show()
```

Histogram yang ditampilkan merupakan representasi visual dari jumlah borough district area di London berdasarkan frekuensi kriminalitasnya. Pada sumbu x, terdapat nilai frekuensi kriminalitas yang diukur dalam jumlah kejadian, sedangkan pada sumbu y, terdapat jumlah borough district area yang terdapat di London.

Dari histogram ini, dapat dilihat distribusi jumlah borough district area di London berdasarkan frekuensi kriminalitasnya. Misalnya, dapat dilihat apakah frekuensi kriminalitas cenderung terpusat pada beberapa borough tertentu atau terdistribusi merata di seluruh borough. Histogram juga dapat membantu mengidentifikasi outlier atau borough district area yang memiliki frekuensi kriminalitas yang jauh lebih tinggi atau rendah dari borough district area lainnya.

Persebaran jumlah kasus antar borough masih cenderung lebih normal dan merata dibandingkan antar lsoa. Nilai tengah menjadi frekuensi terbanyak dan persebaran memusat ke tengah.

# Bar Chart

### 1. Bar Chart - Perbandingan Total Kasus Kriminal Berdasarkan Borough pada tahun 2010

```python
plt.figure(figsize=(12, 10))

# Filter 2010 only
df_2010 = df[df['year'] == 2010]

df_2010.groupby(['borough'])['value'].sum().sort_values(ascending=True).plot(kind='barh', color='#e49393')

plt.title('Perbandingan Total Kasus Kriminal Berdasarkan Borough (Tahun 2010)', fontdict={'fontname':'Arial', 'fontsize': 26, 'fontweight': 'bold', 'verticalalignment': 'bottom', 'y': 2.35}, pad=10)
plt.xlabel('Jumlah Kasus',fontsize=15, fontweight='semibold', fontname='Arial')
plt.ylabel('Frekuensi Kejahatan Kriminal',fontsize=15, fontweight='semibold', fontname='Arial')

plt.show();
```

Dari grafik tersebut, dapat dilihat bahwa borough dengan jumlah kasus kriminal terendah pada tahun 2010 adalah City of London, sementara borough dengan jumlah kasus kriminal tertinggi adalah Westminster. Selain itu, terlihat bahwa terdapat perbedaan yang signifikan dalam jumlah kasus kriminal antara borough-borough yang berbeda di London pada tahun 2010. 

Grafik tersebut dapat digunakan sebagai dasar untuk analisis lebih lanjut tentang pola kriminalitas di London, serta sebagai alat untuk mengidentifikasi daerah-daerah yang memerlukan perhatian khusus dalam upaya penanggulangan kejahatan.

### 2. Bar Chart - Perbandingan Total Kasus Kriminal Berdasarkan Borough pada tahun 2015

Dari grafik tersebut, dapat dilihat bahwa terdapat perbedaan signifikan dalam jumlah kasus kriminal antara borough-borough yang berbeda di London pada tahun 2015.

Grafik ini dapat digunakan sebagai dasar untuk melakukan analisis lebih lanjut tentang pola kriminalitas di London pada tahun 2015, serta sebagai alat untuk mengidentifikasi daerah-daerah yang memerlukan perhatian khusus dalam upaya penanggulangan kejahatan.

# Box Plot

### Box Plot - Distribusi Deskriptif Jumlah Kejadian Kriminal Bulan Setiap Tahun

```python
plt.figure(figsize=(12, 5))

df.groupby(['month', 'year'])[['value']].sum().reset_index().boxplot(by ='year', column=['value'], grid = False);

plt.title('Persebaran Bulanan Tiap Tahun Kriminalitas London Raya', pad=40, 
          fontdict={'fontsize': 18, 'fontweight': 'bold', 'fontname': 'Arial', 
                    'verticalalignment': 'bottom'})
plt.xlabel('Tahun', fontdict={'fontsize': 10, 'fontweight': 'semibold'})
plt.ylabel('Frekuensi Kejahatan Kriminal', fontdict={'fontsize': 10, 'fontweight': 'semibold'})

plt.show();
```

Berdasarkan Boxplot yang terbentuk, dapat dilihat beberapa statistik deskriptif mencakup nilai maksimum, minimum, kuartil, median. Terlihat bahwa median kejadian kriminal bulanan paling tinggi didaptkan pada tahun 2012 dan 2016. Nilai maksimum dan minimum pada tahun tersebut juga ikut naik walaupun pada tahun 2012 terlihat Q1 dan Q3 tampak tidak mengalami kenaikan signifikan.

Dari visualisasi tersebut, dapat disimpulkan juga tidak terdapat bulan pada periode kapan pun yang termasuk outlier.
