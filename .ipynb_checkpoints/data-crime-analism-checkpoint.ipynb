## Deskripsi Permasalahan

Wilayah administrasi London Raya terdiri dari 33 sektor (borough), dimana setiap sektor memiliki pemerintahannya sendiri. Tujuan dari analisis ini adalah untuk mengetahui tindak kriminalitas yang terjadi di Wilayah administrasi London Raya dan setiap sektornya pada rentang bulan Januari 2008 hingga Desember 2016. Dataset yang menjadi dasar analisis memiliki beberapa atribut data yaitu sebagai berikut:

* `lsoa_code`: kode area administrasi dan demografis
* `borough`: sektor dari Wilayah administrasi London Raya
* `major_category`: kategori kriminalitas mayor
* `minor_category`: kategori kriminalitas minor
* `value`: jumlah kejadian tindak kriminalitas sesuai dengan sektor dan waktu (bulan dan tahun)
* `year`: tahun pendataan
* `month`: bulan pendataan

Analisis yang dilakukan didasarkan oleh data dan memiliki tujuan utama untuk menurunkan beberapa kesimpulan yang terlihat dari data keseluruhan dan juga melihat pola atau kejadian menarik yang teramati dalam data.

#### 2. Area Plot - Persebaran Kejahatan Berdasarkan Bulan dan Tahun
Untuk melihat detail persebaran kasus pada tahun 2008 se-wilayah London menggunakan area plot.

```python
df_year_month = df.groupby(['year', 'month']).sum()

# Reset the index to create separate columns for year and month
df_year_month.reset_index(inplace=True)

# Create a pivot table to reshape the data, with months as columns and years as rows
df_pivot = df_year_month.pivot(index='year', columns='month', values='value')

# Create the area plot
df_pivot.plot(kind='area', stacked=False, alpha=0.5, cmap='Paired', figsize=(10, 6))

plt.title('Persebaran Kejahatan Berdasarkan Bulan dan Tahun', 
          fontdict={'fontname':'Arial', 'fontsize': 20, 'fontweight': 'bold', 'verticalalignment': 'bottom', 'y': 1.1})

plt.xticks(rotation=90, fontsize=9, fontname='Arial')
plt.xlabel('Bulan', fontdict={'fontsize': 15, 'fontweight': 'semibold', 'fontname': 'Arial'})
plt.ylabel('Jumlah Kejahatan', fontdict={'fontsize': 15, 'fontweight': 'semibold', 'fontname': 'Arial'})

# Show the plot
plt.show()
```

#### Insight
Berdasarkan visualisasi area plot tersebut, dapat dilihat bahwa terjadi peningkatan kasus kejahatan di wilayah London pada tahun 2011 hingga 2012, dengan puncaknya terjadi pada awal 2012. Setelah itu, kasus kejahatan secara bertahap mengalami penurunan hingga tahun 2014, namun mengalami peningkatan kembali pada tahun 2015. Namun pada tahun 2016, terjadi penurunan kasus kejahatan yang cukup signifikan dibandingkan dengan tahun-tahun sebelumnya.

Selain itu, dapat dilihat juga bahwa kasus kejahatan pada wilayah London memiliki tren musiman, dimana pada bulan-bulan tertentu terjadi peningkatan kasus kejahatan. Bulan Maret hingga Agustus menunjukkan jumlah kasus kejahatan yang relatif tinggi, sedangkan bulan-bulan lainnya menunjukkan jumlah kasus kejahatan yang lebih rendah. Oleh karena itu, pihak yang bertanggung jawab dapat memperhatikan tren musiman ini untuk mengambil langkah-langkah pencegahan yang lebih efektif.

### Word Clouds

#### Word Clouds - Wilayah yang paling sering terjadi kejahatan Major Category 

```python
df_major1 =df[['borough', 'value']].groupby(['borough'], as_index= True).sum()
total =df['value'].sum()
```

```python
# inisiasi wordcloud
max_words = 100
word_string = ''
for borough in df_major1.index.values:
    # memeriksa apakah nama negara adalah nama satu kata
    if len(borough.split(' ')) == 1:
        repeat_num_times = int(df_major1.loc[borough,'value']/float(total)*max_words)
        word_string = word_string + ((borough + ' ') * repeat_num_times)                                    
```

```python
# membuat wordcloud
wordcloud = WordCloud(background_color='white').generate(word_string)
# menampilkan
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(10)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
# buat bentuk london
london_mask = np.array(Image.open('london2.jpg'))
plt.imshow(london_mask,cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
london_wc = WordCloud(background_color='white',mask=london_mask).generate(word_string)
plt.imshow(london_wc)
plt.axis('off')
plt.show()
```

#### Insight
Berdasarkan WordCloud tersebut kejahatan mayor category paling sering terjadi di wilayah Westminster.
